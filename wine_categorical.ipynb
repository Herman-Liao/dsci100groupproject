```r
library(tidyverse)
library(gridExtra)
```

```r
wine <- read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", col_names = FALSE)
```

```r
colnames(wine) <- c('class', 'al', 'acid', 'ash', 'alcalinity', 'mg', 't_phenol',
                    'flava', 'non_f_phenol', 'pro', 'color', 'hue', 'od280',
                    'proline')
wine <- mutate(wine, class = as.factor(class))

options(repr.plot.width = 16, repr.plot.height = 12)
colnames(wine)
```

```r
plot1 <- ggplot(wine, aes(x = hue, y = color, color = class)) +
         geom_point() +
         ggtitle("hue vs color") +
         theme(text = element_text(size = 20))
```

```r
plot2 <- ggplot(wine, aes(x = hue, y = flava, color = class)) +
         geom_point() +
         ggtitle("hue vs flava") +
         theme(text = element_text(size = 20))
```

```r
plot3 <- ggplot(wine, aes(x = al, y = color, color = class)) +
         geom_point() +
         ggtitle("alcohol vs color") +
         theme(text = element_text(size = 20))
```

```r
plot4 <- ggplot(wine, aes(x = al, y = proline, color = class)) +
         geom_point() +
         ggtitle("alcohol vs proline") +
         theme(text = element_text(size = 20))
```

```r
plot5 <- ggplot(wine, aes(x = ash, y = mg, color = class)) +
         geom_point() +
         ggtitle("ash vs mg") +
         theme(text = element_text(size = 20))
```

```r
plot6 <- ggplot(wine, aes(x = alcalinity, y = od280, color = class)) +
         geom_point() +
         ggtitle("alcalinity vs od280") +
         theme(text = element_text(size = 20))
```

```r
plot7 <- ggplot(wine, aes(x = t_phenol, y = non_f_phenol, color = class)) +
         geom_point() +
         ggtitle("t-phenol vs non-f-phenol") +
         theme(text = element_text(size = 20))
```

```r
plot8 <- ggplot(wine, aes(x = al, y = flava, color = class)) +
         geom_point() +
         ggtitle("alcohol vs flava") +
         theme(text = element_text(size = 20))
```

```r
plot9 <- ggplot(wine, aes(x = hue, y = proline, color = class)) +
         geom_point() +
         ggtitle("hue vs proline") +
         theme(text = element_text(size = 20))
```

```r
plot10 <- ggplot(wine, aes(x = hue, y = al, color = class)) +
          geom_point() +
          ggtitle("hue vs alcohol") +
          theme(text = element_text(size = 20))
```

```r
plot11 <- ggplot(wine, aes(x = flava, y = proline, color = class)) +
          geom_point() +
          ggtitle("flava vs. proline") +
          theme(text = element_text(size = 20))
```

```r
grid.arrange(plot2, plot4, plot8, plot9, plot10, plot11, nrow = 3)
```
