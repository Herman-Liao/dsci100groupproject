```r
library(tidyverse)
wine <- read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data", col_names = FALSE)
head(wine)
```

```r
colnames(wine) <- c('class', 'al', 'acid', 'ash', 'alcalinity', 'mg', 't_phenol',
                    'flava', 'non_f_phenol', 'pro', 'color', 'hue', 'od280',
                    'proline')
wine <- mutate(wine, class = as.factor(class))
head(wine)
```

```r
ggplot(wine, aes(x = color, y = hue, color = class)) +
    geom_point() +
    ggtitle("color vs hue")
```

```r
colnames(wine)
ggplot(wine, aes(x = hue, y = flava, color = class)) +
    geom_point() +
    ggtitle("flava vs hue")
```
